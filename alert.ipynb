```python
import pandas as pd
import pandas_ta as ta
import yfinance
```

```python
ticker = yfinance.Ticker("RELIANCE.NS")
df = ticker.history(period="1y")
df
```

```python
adx=df.ta.adx()
rsi=df.ta.rsi()
macd=df.ta.macd()
df=pd.concat([df,adx,rsi,macd],axis=1)
df
```

```python
#Fetching the last row of the df table
last_row=df.iloc[-1]
last_row
```

```python
## Check the trend by conditions on adx
if last_row['ADX_14'] >= 25:
    msg = f"Strong trend : The adx is {last_row['ADX_14']:.2f}"
    print(msg)
if last_row['ADX_14'] < 25:
    msg = f"No trend : The adx is{last_row['ADX_14']}"
    print(msg)
```
