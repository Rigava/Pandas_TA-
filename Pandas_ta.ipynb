## Example of pandas_ta library

```python
# adx=ta.adx(df['high'],df['low'],df['close'])
adx=df.ta.adx()
```

```python
# macd=df.ta.macd()
# macd
```

```python
macd=df.ta.macd(fast=14,slow=28)
# macd
```

```python
rsi=df.ta.rsi()
```
